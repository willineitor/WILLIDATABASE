```python
#calcular tasa de desempleo en hombres y mujeres del quindio
#que se necesita?
#formula TDH = HDQ/PEAHQ -> HDQ: hombres desempleados, 
#desempleados, PEAHQ : hombres en la PEA

#Hombres Quindio

#1 Calcular cantidad de hombres desempleados en el quindio 

HDQ = df_desocupados_genero[(df_desocupados_genero['DSI']== 1) & (df_desocupados_genero['DPTO'] == 63) & (df_desocupados_genero['P3271'] == 1)] ['FEX_C18'].sum()
print(f'Hombres desempleados en el quindio: {HDQ}')
#2. calcular cantidad de hombres en la PEA del quindio 

PEAHQ = df_ft_genero[(df_ft_genero['FT'] == 1) & (df_ft_genero['DPTO'] == 63) & (df_ft_genero['P3271'] == 1)]['FEX_C18'].sum()
print(f'Poblacion economicamente activa hombres en el quindio: {PEAHQ}')

#3. calcular la tasa de desempleo de hombres en el quindio
TDH = HDQ/PEAHQ
print(f'Tasa de desempleo hombres en el quindio: {TDH*100}%')

```
